```python
target = '2000rs11'
data_path = f'/home/rcannon/Code/Radar/{target}/v01/z1500r'
```

```python
logfiles = sorted(glob.glob(f'{data_path}/logfiles/*'))

param = logfiles[0].split('/')[-1].split('_')[0]
print(param)

p_vals = [float(log.split('_')[-1].replace('.log','')) for log in logfiles]
print(p_vals)
```
