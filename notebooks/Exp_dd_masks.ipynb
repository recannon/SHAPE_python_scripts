```python
from astropy.io import fits
import numpy as np
import glob
import matplotlib.pyplot as plt
from astropy import visualization as aviz
```

### 2000WO107

#### Dec01

```python
data_path = f"/home/rcannon/Radar/2000WO107/Goldstone2020/"
night = 'dec01'
set = '6'
#Load delay-Doppler
delay_files = sorted(glob.glob(f"{data_path}/Delay-Doppler/summed/{night}/*20201201*.fits"))

print(len(delay_files))
```

```python
hdu = fits.open(delay_files[0])
data = hdu[0].data

plt.figure()
norm = aviz.ImageNormalize(data,interval=aviz.ZScaleInterval())
plt.imshow(data, norm=norm, origin='lower')
```

```python
mask = np.zeros_like(data)
mask[50:100,70:120] = 1
plt.subplot(1,2,1)
plt.imshow(mask, origin='lower')
masked_data = data*mask
plt.subplot(1,2,2)
plt.imshow(masked_data, norm=norm, origin='lower')
```

```python
for i,dd in enumerate(delay_files):
    hdu = fits.open(dd)
    data = hdu[0].data
    mask = np.zeros_like(data)
    mask[50:100,70:120] = 1
    masked_data = data*mask
    plt.figure()
    plt.imshow(masked_data, norm=norm, origin='lower')
    plt.show()
    plt.close()
    np.savetxt(f'{data_path}/Masks/mask_{set:0>2}_{i:0>2}.txt',mask)
```
