```python
from astropy.io import fits
import numpy as np
import glob
import matplotlib.pyplot as plt
from astropy import visualization as aviz
```

### 2024on

```python
data_path = f"/cephfs/rcannon/2024on/radar/Goldstone2024/"
```

```python

#Load delay-Doppler
cw_files = sorted(glob.glob(f"{data_path}/CW/sep15-44/*sum_1*"))
set = '6'

print(len(cw_files))

for cw in cw_files:
    pow = np.loadtxt(cw)
    plt.plot(pow)
```

```python
mask_i = [120,160]

for i,cw in enumerate(cw_files):
    data = np.loadtxt(cw)
    mask = np.zeros_like(data)
    mask[mask_i[0]:mask_i[1]] = 1
    masked_data = data*mask
    plt.figure()
    plt.plot(data,'k')
    plt.plot(masked_data,'r')
    plt.show()
    plt.close()
    np.savetxt(f'{data_path}/Mask/mask_{set:0>2}_{i:0>2}.txt',mask)
```

#### Nov27

```python
data_path = f"/home/rcannon/Radar/2000WO107/Goldstone2020/"

#Load delay-Doppler
cw_files = sorted(glob.glob(f"{data_path}/CW/summed/nov27/cw*dat"))
set = '0'

print(len(cw_files))
```

```python
for cw in cw_files:
    data = np.loadtxt(cw)[:,0]
    plt.plot(data)
```

```python
plot_i = [4920,5050]
mask_i = [4975,5010]

for i,cw in enumerate(cw_files):
    data = np.loadtxt(cw)[:,0]
    mask = np.zeros_like(data)
    mask[mask_i[0]:mask_i[1]] = 1
    masked_data = data*mask
    plt.figure()
    plt.plot(data[plot_i[0]:plot_i[1]],'k')
    plt.plot(np.arange(mask_i[0]-plot_i[0],mask_i[1]-plot_i[0]),masked_data[mask_i[0]:mask_i[1]],'r')
    plt.show()
    plt.close()
    np.savetxt(f'{data_path}/Masks/mask_{set:0>2}_{i:0>2}.txt',mask)
```
